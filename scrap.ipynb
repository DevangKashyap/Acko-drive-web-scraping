```python
from bs4 import BeautifulSoup
import requests
import pandas as pd
```

```python
website='https://ackodrive.com/collection/hyundai-cars/'
```

```python
response=requests.get(website)
response.status_code
```

```python
Soup = BeautifulSoup(response.content, 'html.parser')
Soup
```

```python
result = Soup.find_all('div', {'data-testid': 'listingcardesktop'})
len(result)
```

```python
name = result[0].find('h2').get_text()
details = result[0].find_all('p')
body_type = details[0].get_text()
seat = details[1].get_text()
variants = details[2].get_text()
fuel = details[3].get_text()
transmission = details[4].get_text()
colour = details[5].get_text()
```

```python
print("Name:",name)
print("colours available:",colour)
print("Body type:",body_type)
print("Seat:",seat)
print("Variants:",variants)
print("Fuel:",fuel)
print("Transmission:",transmission)
```
